Dataset ini berisi mengenai informasi tempat penginapan di Bangkok yang terdaftar di Airbnb.  Terdapat 16 kolom pada dataset 'Airbnb Listings Bangkok', yaitu:

- id: Pengidentifikasi unik untuk setiap properti Airbnb.

## DATA ANALYSIS

Setelah dilakukan tahapan data cleaning kemudian akan dilakukan analisis untuk mencari tahu preferensi kriteria tempat penginapan yang populer di kalangan pelanggan/tamu. Kriteria tempat penginapan yang akan dianalisa terdiri dari beberapa aspek yaitu berupa lokasi, harga sewa, tipe kamar, dan jumlah minimal hari sewa. 


Pada bagian berikut akan ditambahkan kolom berupa data `potential revenue`. Data ini merupakan potensial pendapatan dari satu tempat sewa (ID) dalam satu tahun kedepan. Data kolom ini didapatkan dari hasil kali harga sewa dalam satu malam dengan jumlah hari tersedia untuk disewakan dalam satu tahun.

```python
# Potensial Revenue untuk masing-masing ID

df['potential_revenue'] = df['price']*df['availability_365']
df.head(3)
```

### 1. Lokasi

Berikut ini jumlah kamar untuk setiap area di Bangkok dengan rincian breakdown untuk setiap tipe kamar.

```python
# Proporsi Tipe Kamar dan Totalnya untuk Setiap Area di Bangkok

df_room_per_loc = pd.crosstab(df['neighbourhood'],df['room_type'])
df_room_per_loc['total_room_per_location'] = df_room_per_loc.sum(axis=1)
df_room_per_loc.sort_values('total_room_per_location',ascending=False).head()
```

Dari hasil crosstab antara `neighbourhood` dan `room_type` diatas didapatkan 5 lokasi dengan jumlah kamar sewa terbanyak. Semakin banyak kamar sewa di lokasi tersebut dapat menandakan lokasi tersebut memiliki potensi tinggi untuk dipilih calon tamu untuk menyewa properti karena alasan faktor tertentu berkaitan dengan lokasi tersebut. Namun untuk mencari lokasi mana yang populer lebih kredibel melihat dari parameter jumlah ulasan, oleh karena itu akan dilihat lebih lanjut lokasi mana yang lebih populer dari jumlah ulasan di bagian bawah ini.

```python
# Jumlah Review Total, Jumlah Review Satu Tahun Terakhir, dan Keuntungan Potensial untuk Lima Area di Bangkok dengan Jumla Ulasan Terbanyak dalam 1 Tahun Terakhir

df_loc = df.groupby('neighbourhood').agg({'number_of_reviews': 'sum', 'number_of_reviews_ltm': 'sum','potential_revenue':'sum'}).sort_values('number_of_reviews_ltm',ascending=False).head().reset_index()
df_loc
```

Terdapat perbedaan urutan popularitas lokasi dilihat dari jumlah ulasan total dibandingkan dengan jumlah ulasan 12 bulan terakhir. Penulis akan melihat lokasi terpopuler dari jumlah ulasan 12 bulan terakhir karena data termutakhir akan lebih relevan dan memiliki kesamaan dengan kondisi mendekat di masa yang akan datang saat dijadikan patokan melihat preferensi calon tamu dalam memilih tempat sewa.

Terlihat dari masing-masing tipe kamar memiliki range distribusi harga yang berbeda sehingga pendekatan dalam menganalisa harga harus dilihat dari masing-masing tipe kamar sendiri. Namun agar lebih objektif akan dilakukan Uji Kruskal-Wallis membandingkan apaka rata-rata harga dari setiap tipe kamar sama atau tidak. Berikut statement untuk Ho dan Ha beserta hasil Uji.

Ho : Rata-Rata Harga untuk Setiap Tipe Kamar yang diiklankan di Airbnb untuk Tiga Area Terpopuler Adalah Sama
<br>
Ha : Rata-Rata Harga untuk Setiap Tipe Kamar yang diiklankan di Airbnb untuk Tiga Area Terpopuler Adalah Tidak Sama

```python
# Uji Kruskal-Wallis untuk rata-rata harga setiap tipe kamar di tiga area terpopuler

result = list(kruskal(df[(df['neighbourhood'].isin(list_loc))&(df['room_type']=='Entire home/apt')]['price'], 
        df[(df['neighbourhood'].isin(list_loc))&(df['room_type']=='Private room')]['price'],
        df[(df['neighbourhood'].isin(list_loc))&(df['room_type']=='Hotel room')]['price'],
        df[(df['neighbourhood'].isin(list_loc))&(df['room_type']=='Shared room')]['price']
        ))
print('U-Statistic:', result[0])
print('P-Value:', result[1])

pval =  result[1]

print ('tolak H0')

if pval <= 0.05:
  print('tolak H0')
else:
  print('terima H0')
```

Hasil dari Uji menyatakan untuk tolak Ho yang artinya rata-rata harga untuk setiap tipe kamar yang diiklankan di Airbnb untuk tiga area terpopuler adalah tidak sama sehingga dengan dasar ini lebih meyakinkan untuk melakukan pendekatan dalam menganalisa harga dari masing-masing tipe kamar sendiri.

Pada bagian selanjtnya akan dibandingkan popularitas untuk setiap range harga di masing-masing tipe kamar. Tahap pertama akan dibuat terlebih dahulu range harga di setiap tipe kamar dan sebelumnya harga ekstrim akan dieliminasi.

```python
# Eliminasi outlier price dengan mean std untuk masing-masing tipe kamar

def outlier_mean_std(df, col):
  mean = df[col].mean()
  std = df[col].std()

  upper = df[col].quantile(.95)
  bottom = df[col].quantile(.05)

  return (df[col]<bottom) | (df[col]>upper)

df_normalprice_homeapt = df[(df['neighbourhood'].isin(list_loc))&(df['room_type']=='Entire home/apt')][~outlier_mean_std(df,"price")]
df_normalprice_privateroom = df[(df['neighbourhood'].isin(list_loc))&(df['room_type']=='Private room')][~outlier_mean_std(df,"price")]
df_normalprice_hotelroom = df[(df['neighbourhood'].isin(list_loc))&(df['room_type']=='Hotel room')][~outlier_mean_std(df,"price")]
df_normalprice_sharedroom = df[(df['neighbourhood'].isin(list_loc))&(df['room_type']=='Shared room')][~outlier_mean_std(df,"price")]
```

```python
# Range price

df_normalprice_homeapt['price_equal_freq'] = pd.qcut(x=df_normalprice_homeapt['price'], q=5)
df_normalprice_privateroom['price_equal_freq'] = pd.qcut(x=df_normalprice_privateroom['price'], q=5)
df_normalprice_hotelroom['price_equal_freq'] = pd.qcut(x=df_normalprice_hotelroom['price'], q=5)
df_normalprice_sharedroom['price_equal_freq'] = pd.qcut(x=df_normalprice_sharedroom['price'], q=4)
```

Dibawah ini adalah informasi jumlah ulasan setiap range harga untuk masing-masing tipe kamar

```python
# Jumlah ulasan setiap range harga untuk masing-masing tipe kamar

df_price_homeapt = df_normalprice_homeapt.groupby('price_equal_freq').agg({'number_of_reviews_ltm':('count','sum')}).reset_index().sort_values(('number_of_reviews_ltm','sum'),ascending=False)
df_price_privateroom = df_normalprice_privateroom.groupby('price_equal_freq').agg({'number_of_reviews_ltm':('count','sum')}).reset_index().sort_values(('number_of_reviews_ltm','sum'),ascending=False)
df_price_hotelroom = df_normalprice_hotelroom.groupby('price_equal_freq').agg({'number_of_reviews_ltm':('count','sum')}).reset_index().sort_values(('number_of_reviews_ltm','sum'),ascending=False)
df_price_sharedroom = df_normalprice_sharedroom.groupby('price_equal_freq').agg({'number_of_reviews_ltm':('count','sum')}).reset_index().sort_values(('number_of_reviews_ltm','sum'),ascending=False)

display(df_price_homeapt,df_price_privateroom,df_price_hotelroom,df_price_sharedroom)
```

```python
# Barplot popularitas setiap range harga dari masing-masing tipe kamar

fig, axes = plt.subplots(2, 2, figsize=(10, 8))

axes[0, 0].set_title('Entire Home/Apt')
axes[0, 1].set_title('Private Room')
axes[1, 0].set_title('Hotel Room')
axes[1, 1].set_title('Shared Room')

colors1=['#917FB3','#917FB3','#917FB3','#5C469C','#917FB3']
colors2=['#5C469C','#917FB3','#917FB3','#917FB3','#917FB3']
colors3=['#917FB3','#917FB3','#5C469C','#917FB3','#917FB3']

sns.barplot(ax=axes[0, 0], y=df_price_homeapt['number_of_reviews_ltm','sum'], x=df_price_homeapt['price_equal_freq'],palette=colors1)
sns.barplot(ax=axes[0, 1], y=df_price_privateroom['number_of_reviews_ltm','sum'], x=df_price_privateroom['price_equal_freq'],palette=colors2)
sns.barplot(ax=axes[1, 0], y=df_price_hotelroom['number_of_reviews_ltm','sum'], x=df_price_hotelroom['price_equal_freq'],palette=colors3 )
sns.barplot(ax=axes[1, 1], y=df_price_sharedroom['number_of_reviews_ltm','sum'], x=df_price_sharedroom['price_equal_freq'],palette=colors2 )

for ax in axes.flat:
    for p in ax.patches:
        ax.annotate(f'{int(p.get_height())}', (p.get_x() + p.get_width() / 2., p.get_height()),
                    ha='center', va='center', xytext=(0, 5), textcoords='offset points')

    ax.set_xticklabels(ax.get_xticklabels(), rotation=30)
    ax.set_xlabel('Price Range') 
    ax.set_ylabel('Number of Reviews (LTM)')

plt.subplots_adjust(wspace=0.4, hspace=0.4)


fig.suptitle('Number of Reviews (LTM) by Price Range for each Room Type', fontsize=16,  fontweight='bold')


plt.tight_layout()
plt.show()
```

## Rekomendasi

Dari hasil kesimpulan sebelumnya berikut beberapa rekomendasi yang penulis tawarkan kepada para pemilik properti yang akan atau sudah mengiklankan di Airbnb:

1. Dilihat dari popularitas berdasarkan jumlah ulasan, untuk para pemilik properti direkomendasikan untuk menyediakan tempat sewa akomodasi di area Khlong Toei, Vadhana, dan Ratchatewi. Ketiga area tersebut terletak di pusat Bangkok dan memiliki aksesibilitas yang baik ke berbagai bagian kota. Distrik ini dilayani oleh sistem transportasi yang baik, termasuk stasiun kereta bawah tanah (MRT) dan jaringan jalan utama. Keberadaan transportasi yang mudah dan praktis membuat area ini menjadi lokasi yang menarik bagi wisatawan dan pengunjung yang ingin menjelajahi kota dengan mudah. Selain itu ketiga area ini memiliki daya tarik tersendiri bagi para wisatawan dan pengunjung di area ini. Dengan penjelasan tersebut adalah keputusan yang baik untuk menyediakan tempat sewa akomodasi di sewa tersebut.
2. Tipe kamar entire home/apt jauh lebih unggul secara popularitas berdasarkan jumlah ulasan sehingga tipe kamar ini direkomendasikan untuk menjadi pilihan untuk disewakan bagi para pemilik properti. Tipe kamar shared room sangat jauh tidak populer dibanding tipe kamar lain mengingat banyak tamu cenderung tidak akan nyaman untuk bermalam dalam satu kamar dengan orang lain sehingga tidak begitu disarankan bagi para pemilik properti untuk menyewakan tipe kamar ini.
3. Disarankan untuk para pemilik properti menggunakan kebijakan minimum sewa sebanyak 1 malam. Untuk tipe kamar Entire Home/Apt dan Private Room yang memiliki kriteria cocok untuk sewa jangka panjang dapat memberikan opsi sewa untuk 1 bulan dan memberikan harga spesial untuk menarik calon tamu yang memiliki tujuan sewa akomodasi dalam jangka panjang.
4. Agar memaksimalkan potensi calon tamu memilih kamar sewa disarankan untuk memasang harga di range yang paling populer untuk masing-masing tipe kamar seperti pada detail yang dijelaskan pada bagian kesimpulan. Fasilitas kamar perlu disesuaikan dengan harga yang ditawarkan agar lebih kompetetif dan terlihat lebih menarik dibanding kompetitor sehingga dapat semaksimal mungkin untkuk menarik hati calon tamu saat memutuskan sewa di tempat yang dimiliki pemilik properti.
5. Berkaca dari pemilik properti yang menerima jumlah ulasan banyak dan memiliki keuntungan potensial tinggi, bagi para pemilik properti atau investor yang memiliki budget tinggi disarankan untuk fokus menyediakan tipe kamar Entire Home/Apt agar keuntungan potensial dapat maksimal.
